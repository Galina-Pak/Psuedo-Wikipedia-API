```python
import lib.database_module as db
import lib.wiki_module as wiki
import lib.encoding_module as en
import yaml
```

```python
all_pages_sql = u"""
    SELECT * 
    FROM public.page"""

connection, cursor = db.connect_to_postgres('remote')
cursor.execute(all_pages_sql)
page = cursor.fetchall()
connection.commit()
cursor.close()
connection.close()
column_names = ['page_id','title','created_on','page text']
len(page)
```

```python
def insert_page_vec(page_id,vector):
    add_vector_sql = u"""
        INSERT INTO page_vec (page_id, page_vec)
        VALUES ({}, ARRAY{})
        on conflict (page_id) do 
            UPDATE set page_vec = ARRAY{};
    """.format(page_id, vector,vector)
    connection, cursor = db.connect_to_postgres('remote')
    cursor.execute(add_vector_sql)
    connection.commit()
    cursor.close()
    connection.close()
```

```python
clean_text = []
for p in page:
    clean_text.append(p[3])
clean_ids = []
for p in page:
    clean_ids.append(p[0])
```

```python
transformer = en.build_vectorizer(clean_text)
```

```python
page_vectors = en.get_page_vector(transformer,clean_text,clean_ids)
```

```python
#this uploads all pages vectors to the db
for p in page_vectors:
    insert_page_vec(p,page_vectors[p][0].tolist())
```

```python
def insert_cat_vector(category_id,vector):
    add_category_vector_sql = u"""
        INSERT INTO cate_vec (category_id, category_vec)
        VALUES ({}, ARRAY{})
        on conflict (category_id) do 
            UPDATE set category_vec = ARRAY{};
    """.format(category_id, vector,vector)
    connection, cursor = db.connect_to_postgres('remote')
    cursor.execute(add_category_vector_sql)
    connection.commit()
    cursor.close()
    connection.close()
```

```python
page_cat_dict = {}
for p in page:
    category = db.select_categories_for_page(p[0])[0][1]
    if category in page_cat_dict:
        page_cat_dict[category].append(p[3])
    else:
        page_cat_dict[category] = [p[3]]
```

```python
for p in page_cat_dict:
   page_cat_dict[p] =  "".join(page_cat_dict[p])
```

```python
cat_vectors = []
for p in page_cat_dict:
    cat_vectors.append(en.get_cat_vector(transformer,page_cat_dict[p],p))
```

```python
for cat in cat_vectors:
    insert_cat_vector(category_id=cat.keys()[0],vector=cat[cat.keys()[0]].tolist()[0])
```
